```python
# Dependencies and Setup
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# File to Load
path_2015 = "../Original Data Resources/marathon_results_2015.csv"
path_2016 = "../Original Data Resources/marathon_results_2016.csv"
path_2017 = "../Original Data Resources/marathon_results_2017.csv"

# Read csv Data File and store into Pandas Data Frames
runner_2015_table = pd.read_csv(path_2015)
runner_2016_table = pd.read_csv(path_2016)
runner_2017_table = pd.read_csv(path_2017)
```

```python
# Check all columns for year of 2017
runner_2017_table.columns
```

```python
# Clean up data frame by only keeping necessary columns
runner_2015_df = runner_2015_table[["Bib", "Name", "M/F", "5K", "10K", "15K", 
                                    "20K", "Half", "25K", "30K", "35K", "40K","Pace", 
                                    "Official Time", "Overall"]]
# Add Year column
runner_2015_df["Year"] = "2015"


# Display df
runner_2015_df.head()
```

```python
# Clean up data frame by only keeping necessary columns
runner_2016_df = runner_2016_table[["Bib", "Name", "M/F", "5K", "10K", "15K", 
                                    "20K", "Half", "25K", "30K", "35K", "40K","Pace", 
                                    "Official Time", "Overall"]]
# Add Year column
runner_2016_df["Year"] = "2016"

# Display df
runner_2016_df.head()
```

```python
# Clean up data frame by only keeping necessary columns
runner_2017_df = runner_2017_table[["Bib", "Name", "M/F", "5K", "10K", "15K", 
                                    "20K", "Half", "25K", "30K", "35K", "40K","Pace", 
                                    "Official Time", "Overall"]]
# Add Year column
runner_2017_df["Year"] = "2017"

# Display df
runner_2017_df.head()
```

```python
runner_2017_df = runner_2017_df[(runner_2017_df[["Bib","Name","M/F","5K","10K",
                                                 "15K","20K","25K","30K","35K",
                                                 "40K", "Pace","Official Time", 
                                                 "Overall", "Year"]] != "-").all(axis=1)]
```

```python
runner_2016_df = runner_2016_df[(runner_2016_df[["Bib","Name","M/F","5K","10K",
                                                 "15K","20K","25K","30K","35K",
                                                 "40K", "Pace","Official Time", 
                                                 "Overall", "Year"]] != "-").all(axis=1)]
```

```python
runner_2015_df = runner_2015_df[(runner_2015_df[["Bib","Name","M/F","5K","10K",
                                                 "15K","20K","25K","30K","35K",
                                                 "40K", "Pace","Official Time", 
                                                 "Overall", "Year"]] != "-").all(axis=1)]
```

```python
# List of Data Frame to combine
list_df = [runner_2015_df, runner_2016_df, runner_2017_df]

final_runners_df = pd.concat(list_df)

# Replace blank cell with NaN
final_runners_df.replace("", np.nan, inplace=True)

# Drop NaN
final_runners_df = final_runners_df.dropna(how="any")
```

```python
# Export all Data Frame into csv files
runner_2015_df.to_csv("../Resources/2015_df.csv")
runner_2016_df.to_csv("../Resources/2016_df.csv")
runner_2017_df.to_csv("../Resources/2017_df.csv")
final_runners_df.to_csv("../Resources/final_runners_df.csv")
```
